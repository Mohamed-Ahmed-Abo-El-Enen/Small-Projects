```python
import numpy as np
import os
import time
```

```python
def hms_string(sec_elapsed):
  h = int(sec_elapsed / (60*60))
  m = int((sec_elapsed % (60*60)) / 60) 
  s = sec_elapsed % 60
  return "{} H:{:>02} M:{:>05.2f} S".format(h,m,s)
```

```python
IMAGE_CHANNELS = 3
GENERATE_RESOLUTION_FACTOR = 3
GENERATE_SQUARE = 32 * GENERATE_RESOLUTION_FACTOR

PREVIEW_ROWS, PREVIEW_COLS = 4, 7
PREVIEW_MARGIN = 15

SEED_SIZE = 100

DATA_PATH = "/content/"
FOLDER_NAME = "cartoonset100k"

EPOCHS = 50
BATCH_SIZE = 64
BUFFER_SIZE = 61000

print(f"Image details: {GENERATE_SQUARE}px")
```

```python
training_binary_path = os.path.join(DATA_PATH, f"training_data_{FOLDER_NAME}_{GENERATE_SQUARE}_{GENERATE_SQUARE}.npy")

print("Binary file name: ", training_binary_path)
```

```python
to_folder_index = 3
```

```python
from tqdm import tqdm
import cv2
import glob

def create_save_binary_data_npy(DATA_PATH, FOLDER_NAME, training_binary_path):
  start_time = time.time()
  print("File not found, loading training images...")

  data = []
  faces_path = os.path.join(DATA_PATH, FOLDER_NAME)
  faces_path_file_names = os.listdir(faces_path)
  # *****************************************************************
  faces_path_file_names = faces_path_file_names[:to_folder_index]
  # *****************************************************************
  for file_name in tqdm(faces_path_file_names):
    folder_path = os.path.join(faces_path, file_name)
    image_paths = glob.glob(f"{folder_path}/*.png")
    for image_path in image_paths:
      try:      
        image = cv2.imread(image_path, cv2.IMREAD_COLOR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (GENERATE_SQUARE, GENERATE_SQUARE))
        image = image.astype(np.float32)
        image = (image / 127.5) - 1.0
        data.append(image)

      except Exception as e:
        print(f"ERROR image: {image}, exception: {e}")
        pass

  data = np.array(data)

  print("Saving training image binary file...")
  np.save(training_binary_path, data)
  stop_time = time.time() - start_time
  print(f"Image preprocess & save take: {hms_string(stop_time)}")
  return data
```

```python
def load_binary_data_npy(training_binary_path):
  print("Loading training binary file...")
  start_time = time.time()
  data = np.load(training_binary_path)
  stop_time = time.time() - start_time
  print(f"Load binar y data take: {hms_string(stop_time)}")
  return data
```

```python
if not os.path.isfile(training_binary_path):
  training_data = create_save_binary_data_npy(DATA_PATH, FOLDER_NAME, training_binary_path)
else:
   training_data = load_binary_data_npy(training_binary_path)
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices(training_data).shuffle(buffer_size=BUFFER_SIZE).batch(BATCH_SIZE)
print("Length of total images: ", len(training_data))
print("Length of total batches: ", len(train_dataset))
print("Length of images in one batch: ", len(list(train_dataset.as_numpy_iterator())[0]))
```

```python
from keras.layers import Input, Reshape, Dropout, Dense 
from keras.layers import Flatten, BatchNormalization
from keras.layers import Activation, ZeroPadding2D
from keras.layers import LeakyReLU
from keras.layers import UpSampling2D, Conv2D, Conv2DTranspose
from keras.models import Sequential, Model, load_model
```

```python

def build_generator(seed_size, channels):
  model = Sequential()

  model.add(Dense(units=4*4*256, activation="relu", input_dim=seed_size))
  model.add(BatchNormalization())
  model.add(LeakyReLU())
  model.add(Reshape((4, 4, 256)))

  model.add(Conv2DTranspose(filters=256, kernel_size=(5,5), strides=(2,2), 
                            padding="same", use_bias=False))
  assert model.output_shape == (None, 8, 8, 256)
  model.add(BatchNormalization())
  model.add(LeakyReLU())

  model.add(Conv2DTranspose(filters=256, kernel_size=(5,5), strides=(2,2), 
                            padding="same", use_bias=False))
  assert model.output_shape == (None, 16, 16, 256)
  model.add(BatchNormalization())
  model.add(LeakyReLU())

  model.add(Conv2DTranspose(filters=256, kernel_size=(5,5), strides=(2,2), 
                            padding="same", use_bias=False, activation="tanh"))
  assert model.output_shape == (None, 32, 32, 256)
  model.add(BatchNormalization())
  model.add(LeakyReLU())

  model.add(Conv2DTranspose(filters=64, kernel_size=(5,5), strides=(3,3), 
                            padding="same", use_bias=False, activation="tanh"))
  assert model.output_shape == (None, 96, 96, 64)
  model.add(BatchNormalization())
  model.add(LeakyReLU())

  model.add(Conv2DTranspose(filters=3, kernel_size=(5,5), strides=(1,1), 
                            padding="same", use_bias=False, activation="tanh"))
  assert model.output_shape == (None, 96, 96, 3)

  return model
```

```python
def build_discriminator(image_shape):
  model = Sequential()

  model.add(Conv2D(filters=32, kernel_size=(3,3), strides=(2,2), 
                   padding="same", input_shape=image_shape))
  model.add(LeakyReLU(alpha=0.2))

  model.add(Dropout(rate=0.2))
  model.add(Conv2D(filters=64, kernel_size=(3,3), strides=(2,2), padding="same"))
  model.add(ZeroPadding2D(padding=((0,1), (0,1))))
  model.add(BatchNormalization(momentum=0.7))
  model.add(LeakyReLU(alpha=0.3))

  model.add(Dropout(rate=0.2))
  model.add(Conv2D(filters=128, kernel_size=(3,3), strides=(2,2), padding="same"))
  model.add(BatchNormalization(momentum=0.8))
  model.add(LeakyReLU(alpha=0.2))

  model.add(Dropout(rate=0.2))
  model.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), padding="same"))
  model.add(BatchNormalization(momentum=0.7))
  model.add(LeakyReLU(alpha=0.3))

  model.add(Dropout(rate=0.2))
  model.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1), padding="same"))
  model.add(BatchNormalization(momentum=0.8))
  model.add(LeakyReLU(alpha=0.2))

  model.add(Dropout(rate=0.2))
  model.add(Flatten())
  model.add(Dense(units=1, activation="sigmoid"))

  return model
```

```python
import matplotlib.pyplot as plt

def save_images(generator, cnt, noise, DATA_PATH):
  image_array = np.full((
      PREVIEW_MARGIN + (PREVIEW_ROWS * (GENERATE_SQUARE+PREVIEW_MARGIN)),
      PREVIEW_MARGIN + (PREVIEW_COLS * (GENERATE_SQUARE+PREVIEW_MARGIN)), 3),
      fill_value=255, dtype=np.uint8)
  
  generated_images = generator.predict(noise)
  generated_images = 0.5 * generated_images + 0.5

  image_count = 0 
  for row in range(PREVIEW_ROWS):
    for col in range(PREVIEW_COLS):
      r = row * (GENERATE_SQUARE + 16) + PREVIEW_MARGIN
      c = col * (GENERATE_SQUARE + 16) + PREVIEW_MARGIN
      image_array[r:r+GENERATE_SQUARE, c:c+GENERATE_SQUARE] = generated_images[image_count] * 25
      image_count += 1

  output_path = os.path.join(DATA_PATH, "output")
  if not os.path.exists(output_path):
    os.makedirs(output_path)

  file_name = os.path.join(output_path, f"train_{cnt}.png")
  fig = plt.figure(figsize=(15,15))
  plt.imshow(image_array)
  plt.show()
```

```python
generator = build_generator(SEED_SIZE, IMAGE_CHANNELS)
noise = tf.random.normal([1, SEED_SIZE])
generated_image = generator(noise, training=False)
print("Generated image shape: ", generated_image.shape)

plt.imshow(generated_image[0, :, :, 0])
plt.axis("off")
```

```python
image_shape = (GENERATE_SQUARE, GENERATE_SQUARE, IMAGE_CHANNELS)

discriminator = build_discriminator(image_shape)
decision = discriminator(generated_image)
print(decision)
```

```python
cross_entropy = tf.keras.losses.BinaryCrossentropy()

def generator_loss(fake_output):
  return cross_entropy(tf.ones_like(fake_output), fake_output)
```

```python
def discriminator_loss(real_output, fake_output):
  real_loss = cross_entropy(tf.ones_like(real_output), real_output)
  fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
  total_loss = real_loss + fake_loss
  return total_loss
```

```python
generator_optimizer = tf.keras.optimizers.Adam(1.5e-4, 0.5)
discriminator_optimizer = tf.keras.optimizers.Adam(1.5e-4, 0.5)
```

```python
def generate_save_images(model, epoch, test_input):
  predictions = model(test_input, training=False)

  fig = plt.figure(figsize=(4,4))
  for i in range(predictions.shape[0]):
    plt.subplot(4, 4, i+1)
    plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap="gray")
    plt.axis("off")

  plt.savefig("image_at_epoch{:04d}.png".format(epoch))
  plt.show()
```

```python
@tf.function
def train_step(images):
  seed = tf.random.normal([BATCH_SIZE, SEED_SIZE])

  with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
    generated_images = generator(seed, training=True)

    real_output = discriminator(images, training=True)
    fake_output = discriminator(generated_images, training=True)

    gen_loss = generator_loss(fake_output)
    disc_loss = discriminator_loss(real_output, fake_output)

    generator_gradients = gen_tape.gradient(gen_loss, generator.trainable_variables)
    disriminator_gradients = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(generator_gradients, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(disriminator_gradients, discriminator.trainable_variables))

    #generate_save_images(generator, epoch, seed)
    return gen_loss, disc_loss
```

```python
from IPython import display

def train(dataset, epochs):
  fixed_seed = np.random.normal(0, 1, (PREVIEW_ROWS * PREVIEW_COLS, SEED_SIZE))
  start_time = time.time()

  for epoch in range(epochs):
    epoch_start_time = time.time()

    gen_loss = []
    disc_loss = []
    
    for image_batch in dataset:
      loss = train_step(image_batch)

      gen_loss.append(loss[0])
      disc_loss.append(loss[1])

    g_loss = sum(gen_loss) / len(gen_loss)
    d_loss = sum(disc_loss) / len(disc_loss)

    epoch_stop_time = time.time() - epoch_start_time

    print (f'Epoch {epoch+1}, gen loss={g_loss},disc loss={d_loss}, time={hms_string(epoch_stop_time)}')
    display.clear_output(wait=True)
    save_images(epoch,fixed_seed)

  display.clear_output(wait=True)
  print (f'Epoch {epoch+1}, gen loss={g_loss},disc loss={d_loss}, time={hms_string(epoch_stop_time)}')
  save_images(epoch, fixed_seed)

  stop_time = time.time() - start_time
  print (f'Training time: {hms_string(stop_time)}')
```

```python
train(train_dataset, EPOCHS)
```
